```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('../day_28/train.csv',usecols=['Age','Pclass','SibSp','Parch','Survived'])
df.head()
```

```python
df.dropna(inplace=True)

x=df.drop(columns=['Survived'])
y=df['Survived']

np.mean(cross_val_score(LogisticRegression(),x,y,cv=10,scoring='accuracy'))
```

```python
x['family']=x['SibSp']+x['Parch']+1
x.head()
```

```python
def myfunc(num):
    if num==1:
        return 0
    elif num>1 and num <=4:
        return 1
    else:
        return 2
```

```python
myfunc(3)
```

```python
x['family_type']=x['family'].apply(myfunc)
x.sample(10)
```

```python
x.drop(columns=['SibSp','family','Parch'],inplace=True)
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=10))
```
