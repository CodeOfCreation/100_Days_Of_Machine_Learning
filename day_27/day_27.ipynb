```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("covid_toy.csv")
df.sample(10)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

```python
x_train
```

```python
# simple method

si = SimpleImputer()
x_train_fever=si.fit_transform(x_train[["fever"]])
x_test_fever=si.transform(x_test[['fever']])
x_train.fever.shape
```

```python
# ordinalencoding 

oe=OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough=oe.fit_transform(x_train[["cough"]])
x_test_cough=oe.transform(x_test[['cough']])

x_train_cough.shape
```

```python
# onehotencoding on gender,city

ohe1=OneHotEncoder(drop='first',sparse_output=False)
x_train_gc=ohe1.fit_transform(x_train[['gender','city']])
x_test_gc=ohe1.transform(x_test[['gender','city']])
x_train_gc.shape
```

```python
# Extracting Age
x_train_age = x_train.drop(columns=['gender','fever','cough','city']).values

# also the test data
x_test_age = x_test.drop(columns=['gender','fever','cough','city']).values

x_train_age.shape
```

```python
x_train_transformed=np.concatenate((x_train_age,x_train_fever,x_train_gc,x_train_cough),axis=1)
x_train_transformed.shape
```

```python
# standard method

from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(
    transformers=[
        ('tnf1', SimpleImputer(), ['fever']),
        ('tnf2', OrdinalEncoder(categories=[['Mild', 'Strong']]), ['cough']),
        ('tnf3', OneHotEncoder(drop='first', sparse_output=False), ['gender', 'city'])
    ],
    remainder='passthrough'
)

train_scaled = transformer.fit_transform(x_train)
test_scaled = transformer.transform(x_test)

print(train_scaled.shape)
```
